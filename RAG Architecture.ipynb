```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader

loader = DirectoryLoader('', glob = 'file name (like srt/txt)',show_progress=True, loader_cls=TextLoader)
docs = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
chunks = text_splitter.split_documents(docs)
```

```python
print('Number of Documnets', len(docs))
print('Number of Chunks', len(chunks))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
model_name = ''
embedding_model = HuggingFaceEmbeddings(model_name = model_name)
```

```python
from langchain_chroma import Chroma
```

```python
db = Chroma(collection_name='vector_database',
            embedding_funtion = embedding_model, 
            persist_directory = './chromadb')
```

```python
db.add_documnents(chunks)
```

```python
print(len(db.get()['ids']))
```

Step 2 :Create a Retrieval

```python
query = 'who is the Rachem'

docs_chroma = db.similarity_search_with_score(query, k = 3)
print(len(docs_chroma))
```

```python
context_text = '\n\n'.join([doc.page_content for doc,_score in docs_chroma])

print(context_text)
```

Steps 3: Generation

```python
{'context':context_text, 'question':query}
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_prompt_template = ChatPromptTemplate.from_meassages([(
    'system', '{context}'),
    ('human','{question}')
])

print(chat_prompt_template)
```
